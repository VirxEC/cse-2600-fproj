```python
from pathlib import Path

import joblib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
model: RandomForestClassifier = joblib.load("model-500.joblib")
```

```python
# analyze the model for feature importance
feature_importance = model.feature_importances_
feature_names = model.feature_names_in_
feature_importance = pd.Series(feature_importance, index=feature_names)
feature_importance = feature_importance.sort_values(ascending=False)

# plot the feature importance
plt.figure(figsize=(21, 5))
feature_importance.plot(kind="bar")
plt.title("Feature Importance")
plt.show()
```

```python
cutoff = 0.014

print(len(feature_importance[feature_importance > cutoff]))
print(list(feature_importance[feature_importance > cutoff].keys()))

print(len(feature_importance[feature_importance < cutoff]))
print(list(feature_importance[feature_importance < cutoff].keys()))
```
